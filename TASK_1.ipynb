```python
import matplotlib.pyplot as plt
age_groups = ['0-20 Years', '21-64 Years', '65+ Years']
population_millions = [512, 807, 98]
percentages = [36.1, 57.0, 6.9]
plt.figure(figsize=(8, 5))
bars = plt.bar(age_groups, population_millions, color=['yellow', 'blue', 'pink'])
for i, val in enumerate(population_millions):
    plt.text(i, val + 10, f"{percentages[i]}%", ha='center', fontsize=12)
plt.title("India's Population Distribution by Age (2022)", fontsize=14)
plt.xlabel("Age Group")
plt.ylabel("Population (in millions)")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
